```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set(style = 'darkgrid')
```

```python
s = sns.load_dataset('flights')
s.head()
```

```python
sns.relplot(x = 'passengers', y = 'month', data = s)
```

```python
sns.relplot(x = 'passengers', y = 'month', hue = 'year', data = s)
```

```python
sns.relplot(x = 'passengers', y = 'month', size = 'year', data = s)
```

```python
sns.relplot(x = 'passengers', y = 'month', style = 'year', data = s)
```

```python
b=sns.load_dataset("tips")
b.head()
```

```python
sns.relplot(x='time',y='tip',data=b)
```

```python
sns.relplot(x='time',y='tip',data=b,kind='line')
```

```python
sns.relplot(x='time',y='tip',data=b,kind='scatter')
```

```python
sns.catplot(x='day',y='total_bill',data=b)
```

```python
sns.catplot(x='day',y='total_bill',data=b,kind='violin')
```

```python
sns.catplot(x='day',y='total_bill',data=b,kind='boxen')
```

```python
sns.boxplot(x='day',y='total_bill',data=b)
```

```python
sns.stripplot(x='day',y='total_bill',data=b)
```

```python
sns.swarmplot(x='day',y='total_bill',data=b)
```
